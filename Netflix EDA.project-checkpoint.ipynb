```python
# NETFLIX EDA
# import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the data
df = pd.read_csv(r'C:\Users\.....\Netflix\netflix_titles.csv')
```

```python
df.head(5)
```

```python
# Shape of the dataset
df.shape
```

```python
# Checking the data types
df.dtypes
```

```python
# Checking for missing values
for i in df.columns:
    missing_values = (df[i].isna().sum())
    if missing_values >0:
        print('the missing values for {} is {}'.format(i,missing_values))
```

```python
fig = plt.figure(figsize=(6,6))
plt.plot(df.isna().sum(),color='g')
plt.xticks(rotation='vertical')
plt.title('Missing Values', fontsize=15, fontweight='bold')
plt.show()
```

```python
df.info()
```

```python
# Cheking for duplicates
df['show_id'].duplicated().any()
```

```python
# Changing the date format of the column 'date_added' to 'datetime'
df['date_added']= pd.to_datetime(df['date_added'].str.strip(), format= "%B %d, %Y")
df
```

```python
df.describe()
```

```python
# Calculation of Movies and TV Show in the entire dataset
df['type'].nunique()
print('There are {} types'.format(df['type'].nunique()))
print(df['type'].value_counts())
# Percentage
round(df['type'].value_counts()/ df.shape[0],2)
```

```python
# Visualization of types in the dataset
sns.countplot(data=df,x='type')
plt.ylabel('Value Count')
plt.show()
```

```python
# We can see below that the majority of Netflix's titles have been released in the last decade.

# Distribution of released movies and TV shows
fig = plt.figure(figsize=(12,8),dpi=150)
sns.set(style='darkgrid')
sns.histplot(data=df, x='release_year',hue='type', multiple='stack',bins=30).set(title='Distribution of released movies and TV shows')
plt.show()
```

```python
# Below we can see an increase around 2017 where more titles have been added.
# We can also observe a remarkable differnce between movies and TV shows
```

```python
# Distribution of added movies and TV shows by year
fig = plt.figure(figsize=(12,8),dpi=150)
sns.set(style='darkgrid')
sns.histplot(data=df, x='date_added',hue='type', multiple='stack', bins=30).set(title='Distribution of movies and TV shows added by year')
plt.show()
```

```python
# Distribution of added movies and TV shows by month of the year.

df["Month_date_added"] =df ["date_added"].dt.month.fillna(0)

fig = plt.figure(figsize=(12,8),dpi=150)
sns.set(style='darkgrid')
sns.histplot(data=df, x='Month_date_added',hue='type', multiple='stack',bins=30).set(title='Distribution of added movies and TV shows by month of the year')
plt.show()
```

```python
# Distribution of added movies and TV shows by day of the month.
# In accordance with these data, Netflix uploads the most titles in the first day of month.
df["Day_date_added"] =df ["date_added"].dt.day.fillna(0)
fig = plt.figure(figsize=(12,8),dpi=150)
sns.set(style='darkgrid')
sns.histplot(data=df, x='Day_date_added',hue='type',multiple='stack',bins=30).set(title='Distribution of added movies and TV shows by day of the month')
plt.show()
```
